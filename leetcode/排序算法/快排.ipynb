```python
#快排，首先是要知道怎么随机

import random

#random.random()会随机选择一个0-1之间的浮点数
a = random.random()
print(a)
```

```python
#而random.randint()会随机选择一个整数，1-3都会选到
a = random.randint(1,3)
print(a)
```

```python
array = [4,6,3,9,3,6,7,2,5,7]


def partion(l,r,x):#x是一个数，而不是下标，如果是下标那么就要先存下，否则会随着数组变化去动
    a,x_i = l,0
    for i in range(l,r+1):
        if array[i]<= x:
            array[i],array[a] = array[a], array[i]
            if array[a] == x:
                x_i = a
            a += 1
    array[a-1], array[x_i] = array[x_i], array[a-1]
    return a-1




def quicksort(l,r):
    if l >= r:
        return 
    #首先x是要选择l-r上随机的一个整数
    x = array[random.randint(l,r)]
    mid = partion(l,r,x)
    quicksort(l,mid-1)
    quicksort(mid+1,r)


quicksort(0,len(array)-1)


print(array)


    
```

```python
array = [4,6,3,9,3,6,7,2,5,7]

#进化，荷兰国旗！partion中一次行区分 =，《，》的
def partion(l,r,x):
    a,b,i = l,r,l
    while i <= b:
        if array[i] < x:
            array[i],array[a] = array[a], array[i]
            a +=1
            i +=1
        elif array[i] == x:
            i +=1
        elif array[i] > x:
            array[i],array[b] = array[b], array[i]
            b -= 1
    return a-1,b+1

def quicksort(l,r):
    if l >= r:
        return 
    #首先x是要选择l-r上随机的一个整数
    x = array[random.randint(l,r)]
    a,b = partion(l,r,x)
    quicksort(l,a)
    quicksort(b,r)


quicksort(0,len(array)-1)


print(array)
```
