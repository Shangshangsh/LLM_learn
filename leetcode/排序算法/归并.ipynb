```python
#最小和，核心代码，即把左右两端进行计算
array = [4,6,8,3,5,9]

#这里的m=2，l=0，r = 5
def merge(l,m,r):
    #准备两个变量，一个是i，在l--m上滑动，一个是j，在m+1--r上滑动
    #4  6  8  3  5  9
    #i     m  l
    #如果array[i]<array[ l]说明出现最小和  
    #此时控制一个sum加到最终的ans
    #然后继续，让i往左，j不变，继续滑动，直到找到所有的
    #然后找不到的时候（因为已经排好序了），全部汇总，j继续。
    #所以说，j是要一直控制从m+1到r，因此j要用一个for循环
    #而i则是在while中，当满足条件的时候i才会滑动
    #并且通过一个sum当作中间，即使之前的加过了，sum会保存之前所有j的最小和，可以放心让i划走
    ans = 0
    sum = 0
    i = l
    for j in range(m+1,r+1):
        while i <= m and array[i]<array[j]:
            sum+=array[i]
            i += 1
        ans += sum
    return ans

print(merge(0,2,5))

    






```

```python
#接下来是让左右有序的过程。这个是通用的
array = [4,6,8,3,5,9]
def mergesort(l,m,r):
    #这个的前提是左有序，右也有序，即 4 6 8有序，3 5 9 有序
    #首先需要一个辅助数组来刷，然后i j 控制
    #不断比较，谁小就放到辅助数组里面，然后++，直到一方放完
    #当发现一方放完的时候，把没放完的另一半直接放入到里面就行
    temp = []
    i,j = l,m+1
    #写一个while，不断刷就行，直到一个出界
    while i <= m and j <= r:
        if array[i]<= array[j]:
            temp.append(array[i])
            i += 1
        else:
            temp.append(array[j])
            j += 1

    #当j出界，说明i没出界,剩下的全部塞进去
    while i <= m:
        temp.append(array[i])
        i+=1
    #反之亦然
    while j <= r:
        temp.append(array[j])
        j+=1
    #最后刷回去

    array[l:r+1] = temp
    return array

print(mergesort(0,2,5))


    

```
