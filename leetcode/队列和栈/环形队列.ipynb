回想一下，当我们用一个数组的时候是怎么构造一个队列的，

首先用两个指针，l和r，如果要进来一个，就放在nums[r]处，然后r++，如果要是删除一个，就放在l处，然后l++

现在要求队列长度只能是k，那么逻辑依旧是，准备一个数组，加减和上面说的一样，只不过，当我的r++的时候，需要看一下有没有到达队尾，如果到达队尾了，就跳到0位置就行

那么这样就会出现一个问题，如果数组是满了，（比如一直加，加到最后，l一直没变，r到头后和l重合了，那么之前判断队列为空的逻辑（l==r）就不是了，所以可以直接就维护一个self.size就行了

```python
class MyCircularQueue:
    def __init__(self, k: int):
        self.nums = [None]*k
        self.size = 0
        self.size_max = k
        self.head = 0
        self.tail = 0

    def enQueue(self, value: int) -> bool:
        if self.size < self.size_max:#小于，可以加
            self.nums[self.tail] = value
            if self.tail == self.size_max-1: #如果已经是最后一个了，就到0位置
                self.tail = 0
            else:
                self.tail += 1#如果不是，正常往后移动一位就行
            self.size += 1
            return True
        else:
            return False

    def deQueue(self) -> bool:
        if self.size == 0:#没有元素可以删除
            return False
        else:
            if self.head == self.size_max-1:
                self.head = 0
            else:
                self.head += 1
            self.size -=1
            return True


    def Front(self) -> int:
        if self.size == 0:
            return -1
        else:
            return self.nums[self.head]
        

    def Rear(self) -> int:
        if self.size == 0:
            return -1
        else:
            if self.tail ==0:#这里要跳出-1的位置，但要注意0位置的边缘
                return self.nums[self.size_max-1]
            else:
                return self.nums[self.tail-1]

    def isEmpty(self) -> bool:
        return self.size == 0

    def isFull(self) -> bool:
        return self.size == self.size_max
```
