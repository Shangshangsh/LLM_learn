上面说的都是怎么用数组实现链表，下面说一下怎么用数组实现一个栈

我们都知道，栈是需要先入先出的，那么自然可以想到，当入栈的时候，压入一个，然后++，那么出栈的时候，自然也是从++位置处去出

所以用数组实现一个栈我们只需要维持一个变量 size 就行了

当入元素的时候，入到nums[size]位置，然后size++，让出栈的时候，弹出size-1的位置的元素，然后size--，这样就行了

同时，这个size元素也是栈里元素的数量
