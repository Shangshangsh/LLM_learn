```python
#题目就是，给我两个链表，把他们合并成有序的
class ListNode:
    def __init__(self,value,next):
        self.value = value
        self.next = next
```

```python
#还是首先建立两个链表
head4 = ListNode(12,None)
head3 = ListNode(9,head4)
head2 = ListNode(6,head3)
head = ListNode(3,head2)


head_4 = ListNode(13,None)
head_3 = ListNode(8,head4)
head_2 = ListNode(7,head3)
head_ = ListNode(2,head2)
```

```python
#首先还是先说一下思路，我们需要准备三个变量，一个是pre，一个是c1，一个是c2
#在开始的时候，先比较一下头部，谁小谁就当pre，然后pre.next就可以是c1，另一个头就是c2
#紧接着写一个while循环，当c1c2都不是空时比较，比较c1和c2，谁小，谁就当pre.next,然后pre随之往下移动一个，即pre = pre.next
#那么到最后肯定还有一个没跳没，将那个不是空的处理一下就行

def mergeTwoList(head1:ListNode,head2:ListNode) -> ListNode:
    #首先先看一下两个有没有空
    if head1 == None and head2 != None:
        return head2
    elif head2 == None and head1 != None:
        return head1
    
    #首先定义最后要返回的头部是谁的头部
    head = head1 if head1.value <= head2.value else head2

    c1 = head.next
    #c2是另一个的小的
    c2 = head2 if head == head1 else head1

    pre = head

    while c1 != None and c2 != None:
        if c1.value<=c2.value:
            pre.next = c1
            c1 = c1.next
        else:
            pre.next = c2
            c2 = c2.next
        pre = pre.next

    pre.next = c1 if c1!= None else c2



    return head 
```
