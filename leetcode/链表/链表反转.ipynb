```python
class number:
    def __init__(self,value):
        self.value = value

b = number(5)

print(b.value)
        
```

```python
def g1(b):#这里相当于进去了一个b‘，这个函数是将b’指向了none，所以下面在打印b.value的时候还是5
    b = None

def g2(b):#这里其实还是b‘，但是将b’指向的内存改成了6，由于b和b‘指向的是同一个，所以后面之前的那个b也会变动
    b.value = 6
```

```python

g1(b)
print(b.value)
```

```python

g2(b)
print(b.value)
```

```python
#链表反转，首先定义一个链表的数据结构
class ListNode:
    def __init__(self, value, next):
        self.value = value
        self.next = next
```

```python
#反转链表，准备一个函数,输入的数据是一个listnode
#需要三个变量，head，next，pre。其中head是输入的头节点
# 2 -> 3 -> 4
#head

from re import L


def reverseListNode(head : ListNode) -> ListNode:
    pre = None
    next_ = None
    while head != None:
        next_ = head.next
        head.next = pre
        pre = head
        head = next_

    return pre 
head3 = ListNode(3,None)
head2 = ListNode(2,head3)
head = ListNode(1,head2)
print(head.value, head.next.value, head.next.next.value)
c = reverseListNode(head)
#print(head.value, head.next.value, head.next.next.value)
print(c.value,c.next.value,c.next.next.value)
    


    
```
