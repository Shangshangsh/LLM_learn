这里解释几个概念，python中的%指的是取模（或者叫做取余）

当我们用15%10的时候，10可以除一次那么还剩5，则会输出5

用5%10的时候，10一次也除不了（即除0次），那么依旧是剩5，

所以这样可以用来计算个位

此外，如果需要看是不是进位，则需要整除，比如15//10，这个是用来算能除几次的，除10可以用来算进位

```python
#依旧是需要先建立两个链表
class ListNode:
    def __init__(self,value=0,next=None):
        self.value = value
        self.next = next


head4 = ListNode(9,None)
head3 = ListNode(4,head4)
head2 = ListNode(6,head3)
l2 = ListNode(2,head2)

head_3 = ListNode(9,None)
head_2 = ListNode(4,head_3)
l1 = ListNode(2,head_2)
```

思路首先是，自然可以想到，我们要有一个东西来控制这个进位，而且每次在算的时候要同时计算出这个进位才行

同时，我们可以新建一个链表当最最后的输出

所以思路就是，搞一个类似于双指针的东西，让其滑动，写一个while循环，当两个指针没有同时为空的时候就继续滑动

每次滑倒一个节点，先计算俩人的和（注意这个和可能是两位数），然后计算个位、以及

上一个是不是需要进位

当下算的是不是需要往下进位

因此需要有一个变量统计这个进位

然后一直算一直滑动就行，

最后记得，while走完后，如果还是有进位，那么就要在尾巴上加一个1的节点

```python
def AddTwoListNode(l1:ListNode, l2:ListNode)->ListNode:
    cur = None#这个是用来在我们新建的链表上滑动的
    ans = None#这个是用来在最后返回最终答案的头结点的
    carry = 0 #这个是用来记录进位的
    while l1 != None or l2 != None:
        val1 = l1.value if l1 else 0
        val2 = l2.value if l2 else 0
        cur_sum = val1+ val2 + carry #计算当前两个链表的相加值（包括上一位的进位）
        cur_val = cur_sum % 10 #这里就是余10，即计算个位
        carry = cur_sum//10#这里是整除10，计算进位

        if ans == None:#如果还没开始建表
            ans = ListNode(cur_val)
            cur = ans
        else:
            cur.next = ListNode(cur_val)
            cur = cur.next
        

        l1 = l1.next if l1 else None
        l2 = l2.next if l2 else None
    if carry != 0:
        cur.next = ListNode(carry)
    return ans
fin = AddTwoListNode(l1,l2)
print(fin.value,
      fin.next.value,
      fin.next.next.value,
      fin.next.next.next.value,
      fin.next.next.next.next.value)
    
```
